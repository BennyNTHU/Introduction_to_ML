```python
import pandas as pd

df = pd.DataFrame([['green', 'M', 10.1, 'class2'],
                   ['red', 'L', 13.5, 'class1'],
                   ['blue', 'XL', 15.3, 'class2']])

df.columns = ['color', 'size', 'price', 'classlabel']
df
```

```python
# deal with the ordinal feature
size_mapping = {'XL': 3,'L': 2,'M': 1}  # use dictionary to define a mapping
df['size'] = df['size'].map(size_mapping) # change the size column to numbers
inv_size_mapping = {v: k for k, v in size_mapping.items()} #define the inverse mapping
df['size'].map(inv_size_mapping) # show the inverse mapping

df # show the dataset after mapping
```

```python
# Encoding the labels
import numpy as np

# create a mapping dict to convert class labels from strings to integers
class_mapping = {label: idx for idx, label in enumerate(np.unique(df['classlabel']))}
class_mapping
df['classlabel'] = df['classlabel'].map(class_mapping) # to convert class labels from strings to integers
df

#The inverse mapping
#inv_class_mapping = {v: k for k, v in class_mapping.items()} # reverse the class label mapping
#df['classlabel'] = df['classlabel'].map(inv_class_mapping)
#df
```
